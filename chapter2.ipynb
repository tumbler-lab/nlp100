## 11. タブをスペースに置換
タブ1文字につきスペース1文字に置換せよ．確認にはsedコマンド，trコマンド，もしくはexpandコマンドを用いよ．

```
# プログラム
import re
file_path = 'popular-names.txt'
results = []
with open(file_path, mode='r') as f:
    lines = f.readlines()
    for line in lines:
        tab_to_space = re.sub('\t', ' ', line)
        results.append(tab_to_space)
print(''.join(results))
```

```
# UNIXコマンド
# !man tr
!cat popular-names.txt | tr '\t' ' '
```

## 12. 1列目をcol1.txtに，2列目をcol2.txtに保存
各行の1列目だけを抜き出したものをcol1.txtに，2列目だけを抜き出したものをcol2.txtとしてファイルに保存せよ．確認にはcutコマンドを用いよ．

```
# プログラム
import re
file_path = 'popular-names.txt'
col1 = []
col2 = []
with open(file_path, mode='r') as f:
    lines = f.readlines()
    for line in lines:
        col1.append(line.split('\t')[0] + '\n')
        col2.append(line.split('\t')[1] + '\n')
print(col1)
print(col2)
with open('col1.txt', mode='w') as f:
    f.writelines(col1)
with open('col2.txt', mode='w') as f:
    f.writelines(col2)
```

```
# UNIXコマンド
#! man cut
# -f は区切り文字で区切ったフィールド(列)に対して処理できる(デフォルト区切り文字はタブ)
!cut -f 1 popular-names.txt > col1.txt
!cut -f 2 popular-names.txt > col2.txt
```

## 13. col1.txtとcol2.txtをマージ
12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．

```
# プログラム
file_path = 'popular-names.txt'

results = []
with open('col1.txt', mode='r') as f1:
    with open('col2.txt', mode='r') as f2:
        col1 = f1.readlines()
        col2 = f2.readlines()
        results = [c1.strip('\n') + '\t' + c2 for c1, c2 in zip(col1, col2)]
print(results)
with open('results_13_prog.txt', mode='w') as f:
    f.writelines(results)
with open('results_13_prog.txt', mode='r') as f:
    print(f.read())
```

## 16. ファイルをN分割する

```
# ファイル全体をNこに分けるわけではないようです
# N行ごとに分けるのが正解？
N = int(input())
results = []
with open('popular-names.txt') as f:
    lines = f.readlines()
    len_data = int(len(lines) / N)
    splited_data = []
    for i, line in enumerate(lines):
        splited_data.append(line)
        if  (i+1) % len_data == 0:
            results.append(splited_data)
            splited_data = []
print(results)
```
