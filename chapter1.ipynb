## 04. 元素記号
“Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．

```
letter = 'Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.'
letter = re.sub(r'[,.]', '', letter) # 記号を取る(’,'と'.'の記号を’’に置換する)
words = letter.split(' ')   # 空白で区切ってリストに変換
index_one = [1, 5, 6, 7, 8, 9, 15, 16, 19]  # 先頭の一文字を取り出すインデックス
result = {(word[:1] if index+1 in index_one else word[:2]) : index+1 for index, word  in enumerate(words)}  
# 〇〇〇 if ... else △△△   は...がTrueなら〇〇〇を返す、Falseなら△△△を返すという処理をしている。
# 辞書の内包表示では一回しか":"が使えない？ので上のようにやるしかない？
# result = {word[:1] : index+1 if index+1 in index_one else word[:2] : index+1 for index, word  in enumerate(words)}  はだめ
print(result)
```

## 05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．

```
import re
def get_ngram(letter, n=1, word_unit=True):
    if word_unit:
        inputs = letter.split(' ')
    else :
        inputs = [c for c in re.sub(' ', '', letter)]
    return [inputs[i:n+i] for i in range(0,len(inputs), n)]
 
letter = 'I am an NLPer'
print(get_ngram(letter, n=2))   # 単語単位
print(get_ngram(letter, n=2, word_unit=False))  # 文字単位
```

## 06. 集合
“paraparaparadise”と”paragraph”に含まれる文字bi-gramの集合を，それぞれ, XとYとして求め，XとYの和集合，積集合，差集合を求めよ．さらに，’se’というbi-gramがXおよびYに含まれるかどうかを調べよ．

```
X = get_ngram('paraparaparadise', n=2, word_unit=False)
Y = get_ngram('paragraph', n=2, word_unit=False)
# 失敗
# unhashable type: 'set'
# 中身がリストだと同じものが存在してしまう場合があるのでだめ
# set_X = set(X)
# set_X = {x for x in X}
# set_X = {{xx for xx in x} for x in X}

# 中身がタプルではなく、ジェネレータになってしまうので中身が見れない
# set_X = {(xx for xx in x) for x in X}

set_X = {tuple(xx for xx in x) for x in X}
set_Y = {tuple(yy for yy in y) for y in Y}

print(f'和集合{set_X.union(set_Y)}')
print(f'和集合{set_X | set_Y}')

print(f'積集合{set_X.intersection(set_Y)}')
print(f'積集合{set_X & set_Y}')

print(f'差集合{set_X.difference(set_Y)}')
print(f'差集合{set_X - set_Y}')
```

## 07. テンプレートによる文生成
引数x, y, zを受け取り「x時のyはz」という文字列を返す関数を実装せよ．さらに，x=12, y=”気温”, z=22.4として，実行結果を確認せよ．

```
def templeter(x, y, z):
    # xは時間なので0～24までかつint
    if type(x) is int and 0 <= x <= 12:
        return f'{x}時の{y}は{z}'
    else :
        return f'xは時間ではありません'
print(templeter(12, '気温', 22.4))
print(templeter(1.21, '気温', 22.4))
```

## 08. 暗号文
与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．

英小文字ならば(219 - 文字コード)の文字に置換
その他の文字はそのまま出力
この関数を用い，英語のメッセージを暗号化・復号化せよ．

```
def cipher(plain_text):
    return ''.join([chr(219 - ord(c)) if ord('a') <= ord(c) <= ord('z') else c for c in list(plain_text)])
print(cipher('Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.'))
```

## 09. Typoglycemia
スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え，その実行結果を確認せよ．

```
import random
# ランダムシャッフル
x = list(range(10))
random.shuffle(x)
print(x)

x = list('asdaf')
random.shuffle(x)
print(x)
```

```
text = 'I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .'
result = list()
for word in text.split():
    if len(word) <= 4:
        result.append(word)
        continue
    char_list = list(word[1:-1])
    # print(char_list)
    random.shuffle(char_list)
    word_shuffled = word[0] + ''.join(char_list) + word[-1]
    # print(word_shuffled)
    result.append(word_shuffled)
results = ' '.join(result)
print(results)
```
